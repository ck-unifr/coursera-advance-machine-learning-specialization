## Feature Engineering

```python
def downcast_dtypes(df):
    float_cols = [c for c in df if df[c].dtype == "float64"]
    int_cols   = [c for c in df if df[c].dtype == "int64"]
    
    df[float_cols] = df[float_cols].astype(np.float32)
    df[int_cols]   = df[int_cols].astype(np.int32)
    
    return df
```

```python
# Create "grid" with columns
index_cols = ['shop_id', 'item_id', 'date_block_num']

# For every month we create a grid from all shops/items combinations from that month
grid = [] 
for block_num in df_sales['date_block_num'].unique():
    cur_shops = df_sales.loc[df_sales['date_block_num'] == block_num, 'shop_id'].unique()
    cur_items = df_sales.loc[df_sales['date_block_num'] == block_num, 'item_id'].unique()
    grid.append(np.array(list(product(*[cur_shops, cur_items, [block_num]])),dtype='int32'))

# Turn the grid into a dataframe
grid = pd.DataFrame(np.vstack(grid), columns = index_cols, dtype=np.int32)
grid = pd.concat([grid, df_test])

# Groupby data to get shop-item-month aggregates
gb = df_sales.groupby(index_cols, as_index=False).agg({'item_cnt_day':{'target':'sum'}})
# Fix column names
gb.columns = [col[0] if col[-1]=='' else col[-1] for col in gb.columns.values] 
# Join it to the grid
all_data = pd.merge(grid, gb, how='left', on=index_cols).fillna(0)

# Same as above but with shop-month aggregates
gb = df_sales.groupby(['shop_id', 'date_block_num'],as_index=False).agg({'item_cnt_day':{'target_shop':'sum'}})
gb.columns = [col[0] if col[-1]=='' else col[-1] for col in gb.columns.values]
all_data = pd.merge(all_data, gb, how='left', on=['shop_id', 'date_block_num']).fillna(0)

# Same as above but with item-month aggregates
gb = df_sales.groupby(['item_id', 'date_block_num'],as_index=False).agg({'item_cnt_day':{'target_item':'sum'}})
gb.columns = [col[0] if col[-1] == '' else col[-1] for col in gb.columns.values]
all_data = pd.merge(all_data, gb, how='left', on=['item_id', 'date_block_num']).fillna(0)

# Downcast dtypes from 64 to 32 bit to save memory
all_data = downcast_dtypes(all_data)
del grid, gb 
gc.collect();
```

```python
# List of columns that we will use to create lags
cols_to_rename = list(all_data.columns.difference(['ID'] + index_cols)) 

shift_range = [1, 2, 3, 4, 5, 12]

for month_shift in tqdm_notebook(shift_range):
    train_shift = all_data[index_cols + cols_to_rename].copy()
    
    train_shift['date_block_num'] = train_shift['date_block_num'] + month_shift
    
    foo = lambda x: '{}_lag_{}'.format(x, month_shift) if x in cols_to_rename else x
    train_shift = train_shift.rename(columns=foo)

    all_data = pd.merge(all_data, train_shift, on=index_cols, how='left').fillna(0)

del train_shift

# Don't use old data from year 2013
all_data = all_data[all_data['date_block_num'] >= 12] 

# List of all lagged features
fit_cols = [col for col in all_data.columns if col[-1] in [str(item) for item in shift_range]] 
# We will drop these at fitting stage
to_drop_cols = list(set(list(all_data.columns)) - (set(fit_cols)|set(index_cols))) + ['date_block_num'] 

# Category for each item
# item_category_mapping = items[['item_id','item_category_id']].drop_duplicates()

# all_data = pd.merge(all_data, item_category_mapping, how='left', on='item_id')
all_data = downcast_dtypes(all_data)
gc.collect();
```

### Text feature extraction

```python
feature_cnt = 25

tfidf = feature_extraction.text.TfidfVectorizer(max_features=feature_cnt)
df_items['item_name_len'] = df_items['item_name'].map(len) #Lenth of Item Description
df_items['item_name_wc']  = df_items['item_name'].map(lambda x: len(str(x).split(' '))) #Item Description Word Count
txtFeatures = pd.DataFrame(tfidf.fit_transform(df_items['item_name']).toarray())

cols = txtFeatures.columns
for i in range(feature_cnt):
    df_items['item_name_tfidf_' + str(i)] = txtFeatures[cols[i]]

df_items.drop(columns='item_name', inplace=True)
df_items.head()
```

```python
feature_cnt = 25

tfidf = feature_extraction.text.TfidfVectorizer(max_features=feature_cnt)

df_item_cats['item_category_name_len'] = df_item_cats['item_category_name'].map(len)  #Lenth of Item Category Description
df_item_cats['item_category_name_wc']  = df_item_cats['item_category_name'].map(lambda x: len(str(x).split(' '))) #Item Category Description Word Count

txtFeatures = pd.DataFrame(tfidf.fit_transform(df_item_cats['item_category_name']).toarray())

cols = txtFeatures.columns
for i in range(feature_cnt):
    df_item_cats['item_category_name_tfidf_' + str(i)] = txtFeatures[cols[i]]
    
df_item_cats.drop(columns='item_category_name', inplace=True)
df_item_cats.head()
```

```python
feature_cnt = 25

tfidf = feature_extraction.text.TfidfVectorizer(max_features=feature_cnt)

df_shops['shop_name_len'] = df_shops['shop_name'].map(len)  #Lenth of Shop Name
df_shops['shop_name_wc']  = df_shops['shop_name'].map(lambda x: len(str(x).split(' '))) #Shop Name Word Count

txtFeatures = pd.DataFrame(tfidf.fit_transform(df_shops['shop_name']).toarray())

cols = txtFeatures.columns
for i in range(feature_cnt):
    df_shops['shop_name_tfidf_' + str(i)] = txtFeatures[cols[i]]

df_shops.drop(columns='shop_name', inplace=True)
df_shops.head()
```

```python
#Add items features
all_data = pd.merge(all_data, df_items, how='left', on='item_id')

#Add item category features
all_data = pd.merge(all_data, df_item_cats, how='left', on='item_category_id')

#Add shop features
all_data = pd.merge(all_data, df_shops, how='left', on='shop_id')

all_data = downcast_dtypes(all_data)

all_data.head()
```

### Mean / Variance encoding for category-id last month

```python
mean_enc_item_cat = pd.DataFrame(all_data.groupby(['shop_id', 'item_category_id']).target.agg(['mean', 'var']).reset_index())
mean_enc_item_cat.columns = ['shop_id', 'item_category_id', 'mean_enc_cat_id', 'var_enc_cat_id']

all_data = pd.merge(all_data, mean_enc_item_cat, how='left', on=['shop_id', 'item_category_id'])

del mean_enc_item_cat

all_data = downcast_dtypes(all_data)
```
